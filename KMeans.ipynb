```python
import pandas as pd
```

**Importing CSV**

```python
dados_csv = pd.read_csv("dtframe.csv", encoding="ISO-8859-1", sep=';')

dados_csv
```

```python
tabela_valores = dados_csv[['Português', 'Geografia', 'Física', 'Matemática']]
tabela_valores
```

**Importing KMeans**

```python
from sklearn.cluster import KMeans
```

**Defining 2 clusters and training the model**

```python
grupos_2 = KMeans(n_clusters=2)
grupos_2.fit(tabela_valores)
lista_de_grupos = grupos_2.labels_
lista_de_grupos
```

```python
lista_nomes = dados_csv['Nome']
lista_nomes
```

**Testing grouping**

```python
for i in range(len(tabela_valores)):
  print(i, ' Pertence ao agrupamento N: ', lista_de_grupos[i])
```

```python
for i in range(len(tabela_valores)):
  print(lista_nomes[i], ', na linha N:', i, ' pertence ao agrupamento N: ', lista_de_grupos[i])
```
